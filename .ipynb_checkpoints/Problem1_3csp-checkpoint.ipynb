```python
# problem1.py  –  3-Color Shortest Path (3CSP)
# IT 328 Programming Project 2
#
# NOTE: Portions of this code were written with AI assistance (Claude).
# Specifically: the BFS visited-state key structure (vertex, distance) and
# the num_to_name lookup in read_input_file were suggested by AI.
# All algorithm design, transformation logic, and overall structure
# are the student's own work.

from collections import deque


def read_input_file(filename):
    """
    Parse the 3CSP input file.

    Line 1  : vertices with color suffix, e.g. '1w 2w 3r 4b'
    Lines 2…n+1: n×n adjacency matrix (one row per line)
    Line n+2: 's t'  (bare vertex numbers, no color)
    Line n+3: k
    """
    with open(filename, "r") as f:
        lines = [line.strip() for line in f if line.strip()]

    vertex_tokens = lines[0].split()
    vertex_names = []   # full names like '1w', '2r', ...
    colors = {}         # '1w' -> 'w'
    num_to_name = {}    # '1' -> '1w'   (for resolving bare s/t numbers)

    for token in vertex_tokens:
        vertex_names.append(token)
        colors[token] = token[-1]
        num_to_name[token[:-1]] = token

    n = len(vertex_names)
    adj_matrix = [list(map(int, lines[i].split())) for i in range(1, n + 1)]

    s_raw, t_raw = lines[n + 1].split()
    s = num_to_name[s_raw]
    t = num_to_name[t_raw]
    k = int(lines[n + 2])

    return vertex_names, colors, adj_matrix, s, t, k


def is_valid_next_color(current_color, next_color):
    """
    Enforce the strict 3-color alternation cycle: b → w → r → b → …
    Any other transition is invalid.
    """
    return {"b": "w", "w": "r", "r": "b"}[current_color] == next_color


def solve_3csp(vertex_names, colors, adj_matrix, s, t, k):
    """
    BFS over states (vertex_index, path_length).

    Why (vertex, distance) as the visited key instead of just vertex?
    The same vertex reached at different distances may still lead to t
    via a valid color sequence, so we must not prune based on vertex alone.
    We do cap expansion at distance k to keep the search polynomial.

    Returns (accepted: bool, path: list[str] | None).
    """
    n = len(vertex_names)
    index_map = {vertex_names[i]: i for i in range(n)}
    s_idx = index_map[s]
    t_idx = index_map[t]

    if s == t:
        return True, [s]

    # queue entries: (current_index, distance_so_far, path_so_far)
    queue = deque([(s_idx, 0, [s])])
    visited = {(s_idx, 0)}

    while queue:
        cur, dist, path = queue.popleft()
        if dist >= k:
            continue

        cur_color = colors[vertex_names[cur]]
        for nb in range(n):
            if adj_matrix[cur][nb] != 1:
                continue
            nb_color = colors[vertex_names[nb]]
            if not is_valid_next_color(cur_color, nb_color):
                continue
            new_dist = dist + 1
            state = (nb, new_dist)
            if state in visited:
                continue
            visited.add(state)
            new_path = path + [vertex_names[nb]]
            if nb == t_idx:
                return True, new_path
            queue.append((nb, new_dist, new_path))

    return False, None


def main():
    filename = input("Enter input file name: ")
    try:
        vertex_names, colors, adj_matrix, s, t, k = read_input_file(filename)
        accepted, path = solve_3csp(vertex_names, colors, adj_matrix, s, t, k)
        if accepted:
            print("Accept")
            print("Path   :", " -> ".join(path))
            print("Length :", len(path) - 1)
            print("Colors :", " -> ".join(colors[v] for v in path))
        else:
            print("Reject")
    except FileNotFoundError:
        print("Error: File not found.")
    except Exception as e:
        print("Error:", e)


if __name__ == "__main__":
    main()
```
