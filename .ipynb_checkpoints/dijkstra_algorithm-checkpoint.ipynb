```python
from collections import deque


def read_input_file(filename):
    """
    Reads the BWSP input file and returns:
    - vertex_names: list of vertex labels as strings
    - colors: dictionary mapping vertex label -> color
    - adj_matrix: adjacency matrix as a list of lists of ints
    - s: start vertex label
    - t: target vertex label
    - k: maximum allowed path length
    """
    with open(filename, "r") as file:
        lines = [line.strip() for line in file if line.strip()]

    # First line: vertex labels with colors, like 1w 2w 3r 4b 5b
    vertex_tokens = lines[0].split()

    vertex_names = []
    colors = {}

    for token in vertex_tokens:
        name = token[:-1]     # everything except last character
        color = token[-1]     # last character is the color
        vertex_names.append(name)
        colors[name] = color

    n = len(vertex_names)

    # Next n lines: adjacency matrix
    adj_matrix = []
    for i in range(1, n + 1):
        row = list(map(int, lines[i].split()))
        adj_matrix.append(row)

    # Next line: s t
    s, t = lines[n + 1].split()

    # Final line: k
    k = int(lines[n + 2])

    return vertex_names, colors, adj_matrix, s, t, k


def bfs_bwsp(vertex_names, colors, adj_matrix, s, t, k):
    """
    Solves the Black-White Shortest Path problem using BFS.

    Returns:
    - (True, path) if an acceptable path exists
    - (False, None) otherwise
    """
    # Special case: s == t
    # Path length is 0, so it is valid if k >= 0
    if s == t:
        return True, [s]

    n = len(vertex_names)

    # Map vertex label -> index in adjacency matrix
    index_map = {}
    for i in range(n):
        index_map[vertex_names[i]] = i

    # BFS setup
    queue = deque()
    visited = [False] * n
    parent = [-1] * n
    distance = [-1] * n

    s_index = index_map[s]
    t_index = index_map[t]

    # Start BFS from s
    queue.append(s_index)
    visited[s_index] = True
    distance[s_index] = 0

    while queue:
        current = queue.popleft()

        # Stop exploring if distance already reached k
        if distance[current] == k:
            continue

        for neighbor in range(n):
            # Check if there is an edge
            if adj_matrix[current][neighbor] == 1 and not visited[neighbor]:

                # Extra BWSP rule:
                # the first vertex after s must NOT be red
                if current == s_index and colors[vertex_names[neighbor]] == 'r':
                    continue

                visited[neighbor] = True
                parent[neighbor] = current
                distance[neighbor] = distance[current] + 1
                queue.append(neighbor)

    # If t was not reached or path is too long, reject
    if not visited[t_index] or distance[t_index] > k:
        return False, None

    # Reconstruct path from t back to s
    path = []
    current = t_index
    while current != -1:
        path.append(vertex_names[current])
        current = parent[current]

    path.reverse()
    return True, path


def main():
    filename = input("Enter the input file name: ")

    try:
        vertex_names, colors, adj_matrix, s, t, k = read_input_file(filename)
        accepted, path = bfs_bwsp(vertex_names, colors, adj_matrix, s, t, k)

        if accepted:
            print("Accept")
            print("Path:", " -> ".join(path))
            print("Length:", len(path) - 1)
        else:
            print("Reject")

    except FileNotFoundError:
        print("Error: File not found.")
    except Exception as e:
        print("Error while reading input:", e)


if __name__ == "__main__":
    main()
```
